```python
 from confluent_kafka import Producer
import random
import json
import time

# Kafka producer config
conf = {
    'bootstrap.servers': 'localhost:9092',  # Kafka'nın çalıştığı adres
}

# Producer oluştur
producer = Producer(conf)

# Veri üretme fonksiyonu (örnek veri üretimi)
def generate_data():
    data = {
        'V1': random.random(),
        'V6': random.random(),
        'V8': random.random(),
        'V13': random.random(),
        'V15': random.random(),
        'V19': random.random(),
        'V20': random.random(),
        'V21': random.random(),
        'V23': random.random(),
        'V24': random.random(),
        'V25': random.random(),
        'V26': random.random(),
        'V27': random.random(),
        'V28': random.random(),
        'log_amount': random.random(),
        'Class': random.choice([0, 1])  # 0: Normal, 1: Anomali
    }
    return data

# Veri gönderme döngüsü
def delivery_report(err, msg):
    if err is not None:
        print(f"Mesaj gönderilemedi: {err}")
    else:
        print(f"Mesaj gönderildi: {msg.value().decode('utf-8')}")

# Veri gönderme işlemi
target_count = 100000
sent_count = 0
interval = 1 / 833  # 833 veri/saniye için gerekli interval

while sent_count < target_count:
    data = generate_data()
    producer.produce('input_topic', key=str(data['V1']), value=json.dumps(data), callback=delivery_report)
    producer.poll(0)
    sent_count += 1
    time.sleep(interval)  # 833 veri/saniye için ayar
```
